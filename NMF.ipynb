```python
# ===============================================================
# Set up modules
# ===============================================================

import os

import h5py
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import NMF
```

```python
# ===============================================================
# Import files
# ===============================================================

file_dir='data'
data_csv='Data.csv'
label_csv='Labels.csv'

data_path = os.path.join(file_dir, data_csv)
label_path = os.path.join(file_dir, label_csv)

data_mat = ( np.loadtxt(open(data_path, "rb"), delimiter=";") ).transpose()
label_mat = np.loadtxt(open(label_path, "rb"), delimiter=";")
```

```python
# ===============================================================
# Format and visualize the input matrix
# ===============================================================

my_cmap = plt.cm.hot

#Visualize matrix
data_img = plt.imshow(data_mat, cmap=my_cmap, interpolation='nearest')

plt.show()

```

```python
# Some analysis here

#Set up the NMF model
model = NMF(n_components=4, init='nndsvda')

#Fit model to input matrices
W = model.fit_transform(data_mat)
H = model.components_

```

```python
my_cmap = plt.cm.hot

#Visualize matrix
W_img = plt.imshow(W, cmap=my_cmap, interpolation='nearest')
plt.axes().set_aspect('auto')
#Set figure size
fig = plt.gcf()
fig.set_size_inches(2, 10.5)
#Show figure and colorbar
plt.colorbar()
plt.show()

H_img = plt.imshow(H, cmap=my_cmap, interpolation='nearest')
plt.axes().set_aspect('auto')
#Set figure size
fig = plt.gcf()
fig.set_size_inches(20, 3)
#Show figure and colorbar
plt.colorbar()
plt.show()
```
